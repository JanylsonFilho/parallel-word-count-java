# Análise gráfica dos resultados - Contagem de palavras

Este notebook lê o arquivo `results/resultados.csv`, gerado pelo programa Java, e cria os gráficos exigidos para a análise comparativa entre `SerialCPU`, `ParallelCPU` e `ParallelGPU`.

Fluxo: primeiro execute o Java para gerar o CSV. Depois execute as células deste notebook pelo Jupyter.

```python
from pathlib import Path
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from IPython.display import display
import ipywidgets as widgets

RESULTS = Path('results')
IMAGES = Path('images')
RESULTS.mkdir(exist_ok=True)
IMAGES.mkdir(exist_ok=True)
csv_path = RESULTS / 'resultados.csv'

if not csv_path.exists():
    raise FileNotFoundError('Arquivo results/resultados.csv não encontrado. Execute primeiro o programa Java: ./run.sh the ou run.bat the')

df = pd.read_csv(csv_path)
df = df[df['tempo_ms'] >= 0].copy()
df['arquivo_curto'] = df['arquivo'].str.replace('.txt', '', regex=False)
df['metodo_config'] = df.apply(lambda r: f"{r['metodo']} ({int(r['threads'])} threads)" if r['metodo'] == 'ParallelCPU' else r['metodo'], axis=1)
df.head()
```

## Resumo estatístico
Média, desvio padrão, mínimo e máximo por arquivo, método e configuração de threads.

```python
resumo = (df.groupby(['arquivo', 'tamanho_bytes', 'palavra', 'metodo', 'threads'], as_index=False)
          .agg(ocorrencias=('ocorrencias', 'max'),
               media_ms=('tempo_ms', 'mean'),
               desvio_padrao_ms=('tempo_ms', 'std'),
               minimo_ms=('tempo_ms', 'min'),
               maximo_ms=('tempo_ms', 'max'),
               amostras=('tempo_ms', 'count')))
resumo.to_csv(RESULTS / 'resumo_estatistico.csv', index=False)
display(resumo.sort_values(['arquivo', 'metodo', 'threads']))
```

## Gráfico 1 - Tempo médio por método
Compara o tempo médio de cada estratégia em cada arquivo de entrada.

```python
media_metodo = (df.groupby(['arquivo_curto', 'metodo'], as_index=False)['tempo_ms'].mean()
                .rename(columns={'tempo_ms': 'tempo_medio_ms'}))
fig1 = px.bar(media_metodo, x='arquivo_curto', y='tempo_medio_ms', color='metodo', barmode='group',
              title='Tempo médio por método e arquivo',
              labels={'arquivo_curto': 'Arquivo de entrada', 'tempo_medio_ms': 'Tempo médio (ms)', 'metodo': 'Método'})
fig1.show()
fig1.write_html(IMAGES / 'tempo_medio_por_metodo.html')
try:
    fig1.write_image(IMAGES / 'tempo_medio_por_metodo.png')
except Exception as e:
    print('PNG não exportado. O gráfico interativo HTML foi salvo. Detalhe:', e)
```

## Gráfico 2 - Impacto do tamanho do arquivo
Mostra como o tempo cresce conforme aumenta o tamanho do texto.

```python
media_tamanho = (df.groupby(['arquivo_curto', 'tamanho_bytes', 'metodo'], as_index=False)['tempo_ms'].mean()
                 .rename(columns={'tempo_ms': 'tempo_medio_ms'}))
fig2 = px.line(media_tamanho.sort_values('tamanho_bytes'), x='tamanho_bytes', y='tempo_medio_ms', color='metodo', markers=True,
               hover_data=['arquivo_curto'],
               title='Tempo médio por tamanho do arquivo',
               labels={'tamanho_bytes': 'Tamanho do arquivo (bytes)', 'tempo_medio_ms': 'Tempo médio (ms)', 'metodo': 'Método'})
fig2.show()
fig2.write_html(IMAGES / 'tempo_por_tamanho.html')
try:
    fig2.write_image(IMAGES / 'tempo_por_tamanho.png')
except Exception as e:
    print('PNG não exportado. O gráfico interativo HTML foi salvo. Detalhe:', e)
```

## Gráfico 3 - ParallelCPU por número de threads
Atende ao requisito de investigar diferentes configurações de processamento paralelo.

```python
cpu = df[df['metodo'] == 'ParallelCPU'].copy()
media_threads = (cpu.groupby(['arquivo_curto', 'threads'], as_index=False)['tempo_ms'].mean()
                 .rename(columns={'tempo_ms': 'tempo_medio_ms'}))
fig3 = px.line(media_threads.sort_values('threads'), x='threads', y='tempo_medio_ms', color='arquivo_curto', markers=True,
               title='ParallelCPU: tempo médio por quantidade de threads',
               labels={'threads': 'Número de threads', 'tempo_medio_ms': 'Tempo médio (ms)', 'arquivo_curto': 'Arquivo'})
fig3.show()
fig3.write_html(IMAGES / 'parallel_cpu_threads.html')
try:
    fig3.write_image(IMAGES / 'parallel_cpu_threads.png')
except Exception as e:
    print('PNG não exportado. O gráfico interativo HTML foi salvo. Detalhe:', e)
```

## Gráfico 4 - Speedup
Speedup = tempo médio serial / tempo médio paralelo. Valores maiores que 1 indicam ganho em relação à versão serial.

```python
base = resumo[resumo['metodo'] == 'SerialCPU'][['arquivo', 'media_ms']].rename(columns={'media_ms': 'serial_ms'})
paralelos = resumo[resumo['metodo'].isin(['ParallelCPU', 'ParallelGPU'])].merge(base, on='arquivo', how='left')
paralelos['speedup'] = paralelos['serial_ms'] / paralelos['media_ms']
paralelos['configuracao'] = paralelos.apply(lambda r: f"{r['metodo']} ({int(r['threads'])} threads)" if r['metodo'] == 'ParallelCPU' else r['metodo'], axis=1)
paralelos.to_csv(RESULTS / 'speedup.csv', index=False)
fig4 = px.bar(paralelos, x='arquivo', y='speedup', color='configuracao', barmode='group',
              title='Speedup das versões paralelas em relação à SerialCPU',
              labels={'arquivo': 'Arquivo', 'speedup': 'Speedup', 'configuracao': 'Configuração'})
fig4.add_hline(y=1, line_dash='dash')
fig4.show()
fig4.write_html(IMAGES / 'speedup.html')
try:
    fig4.write_image(IMAGES / 'speedup.png')
except Exception as e:
    print('PNG não exportado. O gráfico interativo HTML foi salvo. Detalhe:', e)
```

## Interface interativa no Jupyter
Use os seletores abaixo para filtrar arquivo e método. Este trecho reforça o uso do Jupyter como interface gráfica de análise.

```python
arquivo_widget = widgets.Dropdown(options=sorted(df['arquivo'].unique()), description='Arquivo:')
metodo_widget = widgets.SelectMultiple(options=sorted(df['metodo'].unique()), value=tuple(sorted(df['metodo'].unique())), description='Métodos:')

def atualizar(arquivo, metodos):
    filtrado = df[(df['arquivo'] == arquivo) & (df['metodo'].isin(list(metodos)))]
    fig = px.box(filtrado, x='metodo_config', y='tempo_ms', points='all',
                 title=f'Distribuição dos tempos - {arquivo}',
                 labels={'metodo_config': 'Método/configuração', 'tempo_ms': 'Tempo (ms)'})
    fig.show()

widgets.interact(atualizar, arquivo=arquivo_widget, metodos=metodo_widget)
```
